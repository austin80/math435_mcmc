```python
def colorful_vertex_plot(g, pos, attr, node_size = 75, cmap = plt.cm.jet, plot_title = ''):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
    ax.set_title(plot_title)
    ax.axis('Off')
    groups = set(nx.get_node_attributes(g, attr).values())
    mapping = dict(zip(sorted(groups),count()))
    nodes = g.nodes()
    colors = [mapping[g.node[n][attr]] for n in nodes]
    ec = nx.draw_networkx_edges(g, pos, alpha=0.2)
    nc = nx.draw_networkx_nodes(g, pos, nodelist=nodes, node_color=colors, with_labels=False, node_size = node_size, cmap = cmap)

    #cbar = fig.colorbar(mappable=nc, ax=ax)
    #cbar.set_label(attr)
    #cbar.set_ticks([min(colors),max(colors)])
    #cbar.set_ticklabels(['low','high'])
    #plt.savefig('COprecincts.png', bbox_inches='tight', pad_inches=0, transparent = True)
```

```python
def is_valid_swap(G, x, y):
    x_dist = G.node[x]['district']
    y_dist = G.node[y]['district']
    Dx = G.subgraph([node for node in G.nodes if G.node[node]['district'] == x_dist and not node == x])
    Dy = G.subgraph([node for node in G.nodes if G.node[node]['district'] == y_dist and not node == y])
    for comp in nx.connected_components(Dx):
        if set(G.neighbors(y)).intersection(set(comp))== set():
            return False
    for comp in nx.connected_components(Dy):
        if set(G.neighbors(x)).intersection(set(comp))== set():
            return False
    return True

def update_swaps(G, x, y):
    nodes_to_check = [x]+[y]+list(G.neighbors(x))+list(G.neighbors(y))
    for node in G.neighbors(x):
        for nbr in G.neighbors(node):
            nodes_to_check.append(nbr)
    for node in G.neighbors(y):
        for nbr in G.neighbors(node):
            nodes_to_check.append(nbr)
    nodes_to_check = list(set(nodes_to_check))
    x_dist = G.nodes[x]['district']
    y_dist = G.nodes[y]['district']
    for z in nodes_to_check:
        G.nodes[z]['swaps'] = []
        dists_to_check = list(set([x_dist, y_dist]+[G.node[nbr]['district'] for nbr in G.neighbors(z)]))
        for dist in dists_to_check:
            if G.nodes[z]['district'] == dist:
                pass
            else:
                dist_nodes = [node for node in G.nodes if G.nodes[node]['district']==dist]
                for node in dist_nodes:
                    if is_valid_swap(G, z, node):
                        G.nodes[z]['swaps'].append(node)
                        if z not in G.nodes[node]['swaps']:
                            G.nodes[node]['swaps'].append(z)
                    else:
                        if z in G.nodes[node]['swaps']:
                            G.nodes[node]['swaps'].remove(z)
                            
#     for z in nodes_to_check:
#         G.nodes[z]['swaps'] = []
#         for node in nodes:
#             if G.nodes[node]['district'] == G.nodes[z]['district']:
#                 pass
#             else:
#                 if is_valid_swap(G, z, node):
#                     G.nodes[z]['swaps'].append(node)
#                     if z not in G.nodes[node]['swaps']:
#                         G.nodes[node]['swaps'].append(z)
#                 else:
#                     if z in G.nodes[node]['swaps']:
#                         G.nodes[node]['swaps'].remove(z)

# def update_seats(seats, x, y):
#     if G.node[x]['party'] == 1:
#         seats[G.node[x]['district']]+=1
#         seats[G.node[y]['district']]-=1
#     if G.node[y]['party'] == 1:
#         seats[G.node[y]['district']]+=1
#         seats[G.node[x]['district']]-=1
#     return seats
```

```python
G = nx.grid_graph(dim=[15,15])
for i in range(0,15):
    for j in range (0,15):
        G.nodes[(i,j)]['district']=i
        G.nodes[(i,j)]['party']=1
        if j==0:
            G.node[(i,j)]['party']=0
        G.node[(i,j)]['swaps']=[]
pos = dict((node, node) for node in G.nodes)
nodes = list(G.nodes)
for i in range(len(nodes)):
    for j in range(i, len(nodes)):
        if G.nodes[nodes[i]]['district']==G.nodes[nodes[j]]['district']:
            pass
        else:
            if is_valid_swap(G, nodes[i],nodes[j]):
                G.nodes[nodes[i]]['swaps'].append(nodes[j])
                G.nodes[nodes[j]]['swaps'].append(nodes[i])
```

```python
N = 1000
t = 0
while t<N:
    per = (t/N)*100
    sys.stdout.write("\r%f%%" %per)
    sys.stdout.flush()
    proposal = list(random.choice(swaps))
    x = proposal[0]
    y = proposal[1]
    proposal_G = copy.deepcopy(G)
    x_dist = proposal_G.nodes[x]['district']
    y_dist = proposal_G.nodes[y]['district']
    proposal_G.nodes[x]['district'] = y_dist
    proposal_G.nodes[y]['district'] = x_dist
    update_swaps(proposal_G, x, y)
    new_swaps = []
    for node in nodes:
        for swap in proposal_G.nodes[node]['swaps']:
            if {node, swap} not in new_swaps:
                new_swaps.append({node, swap})
    p = min([len(swaps)/len(new_swaps),1])
    coin = np.random.binomial(1,p)
    if coin == 1:
        G = proposal_G
        swaps = new_swaps
        t+=1
    else:
        t+=1
```

```python
def is_boundary_node(node):
    dist = G.node[node]['district']
    nbrs = [nbr for nbr in G.neighbors(node)]
    for nbr in nbrs:
        if not G.node[nbr]['district']==dist:
            return True
    return False

def all_pairs_discrepancy(dist_sizes):
    val = 0
    for i in range(0,len(dist_sizes)):
        for j in range(i,len(dist_sizes)):
            val+=abs(dist_sizes[i]-dist_sizes[j])
    return val

def max_discrepancy(dist_sizes):
    dist_sizes.sort()
    return dist_sizes[-1]-dist_sizes[0]

def unif_TVD(dist_sizes):
    S = sum(dist_sizes)
    U = S/len(dist_sizes)
    unif_dist_sizes = [1/U]*len(dist_sizes)
    return sum(abs((1/S)*x-(1/U)) for x in dist_sizes)
```

```python
b = 55
N = 10000
t=0
boundary_nodes = [node for node in G.nodes if is_boundary_node(node)]
dist_sizes = [len([node for node in G.nodes if G.node[node]['district']==i]) for i in range(0,10)]
while t<N:
    node_to_change = random.choice(boundary_nodes)
    dist = G.node[node_to_change]['district']
    G.node[node_to_change]['district']=None
    dist_graph = G.subgraph([node for node in G.nodes if G.node[node]['district']==dist])
    if len(dist_graph)==0:
        G.node[node_to_change]['district']=dist
        t+=1
    elif nx.is_connected(dist_graph):
        neighbor_districts = list(set(G.node[nbr]['district'] for nbr in G.neighbors(node_to_change) if not G.node[nbr]['district']==dist))
        forward_prob = (1/len(boundary_nodes))*(1/len(neighbor_districts))*math.e**(-b*unif_TVD(dist_sizes))
        new_dist = random.choice(neighbor_districts)
        if not dist_sizes[new_dist]==1:
            G.node[node_to_change]['district']=new_dist
            new_neighbor_districts = [x for x in neighbor_districts if not x==new_dist]+[dist]
            new_boundary_nodes = [node for node in boundary_nodes if node not in G.neighbors(node_to_change)]+[node for node in G.neighbors(node_to_change) if is_boundary_node(node)]
            new_dist_sizes = []
            for i, x in enumerate(dist_sizes):
                if i==dist:
                    new_dist_sizes.append(x-1)
                elif i==new_dist:
                    new_dist_sizes.append(x+1)
                else:
                    new_dist_sizes.append(x)
            backward_prob = (1/len(new_boundary_nodes))*(1/len(new_neighbor_districts))*math.e**(-b*unif_TVD(new_dist_sizes))
            p = min(backward_prob/forward_prob,1)
            coin = np.random.binomial(1,p)
            if coin == 1:
                boundary_nodes = new_boundary_nodes
                dist_sizes = new_dist_sizes
            else:
                G.node[node_to_change]['district'] = dist
        else:
            G.node[node_to_change]['district'] = dist
        t+=1
    else:
        G.node[node_to_change]['district'] = dist
        t+=1
```
